```java
public class Car{
    protected String brand;
    protected String wheelType;
    protected static int carCount;

    public Car(String brand, String wheelType){
        this.brand = brand;
        this.wheelType = wheelType;
        Car.carCount++;
    }

    public static void main(String[] args){
        Car car1 = new Car("Toyota", "Alloy");
        Car car2 = new Car("Lamborghini", "Steel");

        System.out.println("Car - " + car1.brand + "Wheel - " + car1.wheelType);
        System.out.println("Car - " + car2.brand + "Wheel - " + car2.wheelType);
        System.out.println("Number of Cars - " + Car.carCount);
    }
}

Car.main(null);
```
